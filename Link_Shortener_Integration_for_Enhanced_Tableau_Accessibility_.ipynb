```python
!pip install pandas wordcloud

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Text data
text = "Today, I was named a #Tableau Ambassador. I had the good fortune of being introduced to Tableau in 2013, and fell in love with its potential to drive positive change in organizations and the lives of individuals that use it. My sincerest gratitude to Tableau & Salesforce for this honor. And thanks to so many colleagues and community friends that have contributed significantly to my journey with Tableau & analytics."

# Create a WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the WordCloud (optional)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Save the WordCloud image to a file
wordcloud.to_file('wordcloud.png')

# Export word frequencies to a CSV file (optional)
word_freq = WordCloud(width=800, height=400, background_color='white').process_text(text)
df = pd.DataFrame(list(word_freq.items()), columns=['Word', 'Frequency'])
df.to_csv('word_frequencies.csv', index=False)
```
