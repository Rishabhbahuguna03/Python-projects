```python
import matplotlib.pyplot as plt
import datetime

# Sample data - replace this with your own data
dates = ["2023-09-30", "2023-10-01", "2023-10-02", "2023-10-03", "2023-10-04", "2023-10-05", "2023-10-06"]
study_hours = [1, 1.5, 2, 1.25, 1.75, 0, 0]  # Make sure this list matches the number of dates

# Convert dates to datetime objects
dates = [datetime.datetime.strptime(date, "%Y-%m-%d") for date in dates]

# Create a figure and axis
plt.figure(figsize=(10, 6))
plt.plot(dates, study_hours, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Hours Studied')
plt.title('Spanish Learning Progress')

# Format x-axis to display dates nicely
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()

```
